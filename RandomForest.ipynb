```python
import pandas as pd
import numpy as np
import sklearn.tree as tree
import sklearn.model_selection as ms
import sklearn.ensemble as ens
import sklearn.metrics as met
rf = ens. RandomForestClassifier(n_estimators=100)
```

```python
DATASET
```

```python
data = pd.read_csv('Desktop/panca/KECERDASAN BISNIS/adult_data.csv')
```

```python
data.head(10)
```

```python
CLEANING DATA UNTUK INPUTAN BERNILAI 0
```

```python
data.dropna(inplace=True)
```

```python
PREPROCESSING DATA DENGAN LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
```

```python
data['sex'] = enc.fit_transform(data['sex'].values)
data['education'] = enc.fit_transform(data['education'].values)
data['race'] = enc.fit_transform(data['race'].values)
data['marital_status'] = enc.fit_transform(data['marital_status'].values)
data['workclass'] = enc.fit_transform(data['workclass'].values)
data['occupation'] = enc.fit_transform(data['occupation'].values)
data['relationship'] = enc.fit_transform(data['relationship'].values)
data['native'] = enc.fit_transform(data['native'].values)
```

```python
X = data.drop(['fnlwft', 'workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'income'], axis=1)
y = data['income']
```

```python
MODELING DATA
```

```python
X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size = 0.3)
```

```python
rf.fit(X_train, y_train)
y_prediksi = rf.predict(X_test)
print('Akurasi = ', met.accuracy_score(y_test, y_prediksi))
print(met.classification_report(y_test, y_prediksi))
print(rf.feature_importances_)
```
